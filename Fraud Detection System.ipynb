# Fraud Detection System
Detect fraudulent transactions using Isolation Forest.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report

# Load data
df = pd.read_csv("/content/drive/MyDrive/PPP/creditcard_fraud_clean.csv")

# Separate features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model: Isolation Forest
model = IsolationForest(contamination=0.001)
model.fit(X_train)

# Predict
y_pred = model.predict(X_test)
y_pred = [1 if val == -1 else 0 for val in y_pred]  # Flip anomaly labels

# Evaluation
print(classification_report(y_test, y_pred))
```
